```python
x=df.drop("Outcome",axis=1)
```

```python
y=df.Outcome
```

```python
from sklearn.ensemble import ExtraTreesClassifier
model = ExtraTreesClassifier()
model.fit(x,y)
```

```python
print(model.feature_importances_)
```

```python
##plot graph for feature importance for better visualisation

feat_importance=pd.Series(model.feature_importances_,index=x.columns)
feat_importance.nlargest(5).plot(kind='barh')
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.svm import SVC
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
X_train
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
X_train
```

```python
lr =LogisticRegression()
lr.fit(X_train,y_train)
print(f"the training score of the logistic regression is :{lr.score(X_train,y_train)*100}")
```

```python
lr_pred = lr.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
print("the accuracy score of the logistic regression is",accuracy_score(lr_pred,y_test)*100)
```

KNN classifier

```python
knn = KNeighborsClassifier(n_neighbors=5)
```

```python
knn.fit(X_train,y_train)
```

```python
print(f"the training score of the knn is :{knn.score(X_train,y_train)*100}")
```

```python
knn_pred = knn.predict(X_test)
```

```python
print("the accuracy score of the knn is",accuracy_score(knn_pred,y_test)*100)
```

Random Forest Classifier

```python
rf = RandomForestClassifier()
rf.fit(X_train,y_train)
```

```python
print(f"the training score of the Random Forest is :{rf.score(X_train,y_train)*100}")
```

```python
rf_pred = rf.predict(X_test)
```

```python
print("the accuracy score of the Random forest is",accuracy_score(rf_pred,y_test)*100)
```

XGboost

```python
xgb = XGBClassifier()
xgb.fit(X_train,y_train)
```

```python
print(f"the training score of the Random Forest is :{xgb.score(X_train,y_train)*100}")
```

```python
xgb_pred = xgb.predict(X_test)
```

```python
print("the accuracy score of the knn is",accuracy_score(xgb_pred,y_test)*100)
```
